```python
# Задание 1 Создать кортеж с числами от 5 до 9, затем создать словарь,
# где ключами будут элементы этого кортежа, а значениями - списки, 
# содержащие числа от 0 до значения текущего ключа.
# Напечатать пары ключ-значение данного словаря, пронумеровав их

myTup = tuple(range(5, 10))
myDict = {}

#Инициализируем словарь
for tupItem in myTup:
    myDict[tupItem] = list(range(0, tupItem + 1))

#Выводим содержимое словаря
n = 1
for key, value in myDict.items():
    print("Пара ключ-значение номер {0}. Ключ: {1}, значение: {2}. \n".format(n, key, value))
    n += 1
```

```python
# Задание 2. Создать переменную строкового типа с вашими ФИО. 
# Посчитать количество вхождений в эту строку буквы "а",
# заменить каждое вхождение на символ "@".
# Разбить исходную строку по пробелам в список.
# С помощью функции строк format() напечатать "Меня зовут ~имя", 
# используя обращение к списку из прошлого предложения.

fullName = "Власенко Дмитрий Андреевич"
print("Количество вхождений буквы \"а\" в мое ФИО: {}\n".format(fullName.count("а")))

changedName = fullName.replace("а", "@")
print("Имя после замены символов: {}\n".format(changedName))

nameList = fullName.split(" ")
print("Меня зовут {}".format(nameList[1]))
```

```python
# Задание 3. Установить модуль itertools. Если будет ошибка,
# что такого нет, то попробуйте more-itertools.
# Найти перестановки из списка с ФИО Задания 2.

import itertools as it

permutationList = list(it.permutations(nameList))
print("Все перестановки из списка с ФИО:\n\n {}".format(permutationList))
```

```python
#Задание 4. Вычислить e^(sin(4^2 - ln(85)))

import math

result = math.exp(math.sin(4**2 - math.log(85)))
print("Полученный результат: {}".format(result))
```

```python
# Задание 5. Создать 2 двумерных массива (матрицы) произвольного размера 
# (главное, чтобы было одинаковое количество строк),
# в одном из которых все значения - единицы, а в другом - единицы
# на главной диагонали, а остальные значения нулевые.
# Соедините эти матрицы в одну так, чтобы число строк
# осталось исходным, а число столбцов стало суммарным.

import numpy as np

rowCount = 5
firstMatrix = np.ones((rowCount, 9)) # матрица из единиц
print("Матрица 1: \n\n{}\n".format(firstMatrix))

secondMatrix = np.eye(rowCount, rowCount) # матрица с единицами по главной диагонали
print("Матрица 2: \n\n{}\n".format(secondMatrix))

result = np.concatenate((firstMatrix, secondMatrix), axis = 1)
print("Итоговая матрица: \n\n{}".format(result))
```
