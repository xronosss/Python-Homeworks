```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR
```

```python
# Подготовка данных
x = np.arange(100)
y = 2.2 * x + 4
noise = 25 * (np.random.rand(100)-0.5)
y += noise
lr = LR() 
lr = lr.fit(x.reshape((-1, 1)), y) 

# Задание 1

regResidual = y - lr.predict(x.reshape(-1, 1))
plt.rcParams['figure.figsize'] = (16, 8)
plt.hist(regResidual)
plt.title('Диаграмма распределения остатков', fontsize=25)
plt.xlabel('Интервал', fontsize=25)
plt.ylabel('Количество попаданий в интервал', fontsize=25)
plt.grid()
plt.show()
```

```python
# Задание 2

# Подготовка данных
x = np.arange(0, 10, 0.1)
y = np.sin(x)
noise = np.random.rand(100) * 0.3
y += noise

# Визуализация
plt.rcParams['figure.figsize'] = (16, 8)
plt.scatter(x, y)
plt.title('Синусоида с шумом', fontsize=25)
plt.ylabel('Y', fontsize=25)
plt.xlabel('X', fontsize=25)
plt.show()

# Корреляция
corrCoefArray = []
for i in range(1, 101):
    corrCoef = np.corrcoef(y, np.roll(y, i))[0,1]
    corrCoefArray.append(corrCoef)
plt.rcParams['figure.figsize'] = (16, 8)
plt.scatter(range(1, 101), corrCoefArray, color='red')
plt.title('Корреляции сдвига', fontsize=25)
plt.ylabel('Коэффициент корреляции', fontsize=25)
plt.xlabel('Сдвиг', fontsize=25)
plt.show()
```
