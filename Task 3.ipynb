```
# Подключаем библиотеки

from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
# ЗАДАНИЕ 1

# Загружаем данные

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Строим графики
data = [X[y==0, 2], X[y==1, 2], X[y==2, 2]]
plt.rcParams['figure.figsize'] = (20, 7)
plt.title('График boxplot', fontsize=15)
plt.ylabel('Значение третьего признака', fontsize=18)
plt.xlabel('Номер класса', fontsize=15)
plt.boxplot(data)
plt.grid()
plt.show()
```

```
# ЗАДАНИЕ 2

# Создадим две выборки
zeroSample = X[y==0, 2][:24]
firstSample = X[y==1, 2][:24]

# Посчитаем дисперсии выборок
var0 = np.var(zeroSample)
var1 = np.var(firstSample)
print('Дисперсия нулевой выборки: {0}, дисперсия первой выборки: {1}'.format(var0, var1))

# Посчитаем F-критерий
import scipy.stats as stats
print(stats.f_oneway(firstSample,zeroSample))
```
